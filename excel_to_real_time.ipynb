#Convert Any Excel Sheet To Real-Time Data And Build Power BI Dashboard

##Problem Statement:
If you have data in an Excel file and you want to make an automatically refreshing dashboard like the one in the video, read this notebook.<br>
In the real-working, one the main goals of the dashboard is monitoring.<br>
Sales monitoring, Cost monitoring, operations monitoring...etc.<br>
For example, the dashboard in the video monitors the sales of a pizza restaurant (every time a new customer comes and orders a pizza, the sales values ​​in the dashboard change automatically).<br>
When you are just beginning to learn data analysis, it is difficult to find real-time data,
So I developed a Python code(available Windows app) that converts any Excel sheet into dynamic data. Even if you have no experience with Python, you can easily apply it to your data  if you read the instructions.<br>
In this notebook, I will put the Python code/App and how to use it, along with the example of the dashboard in the video.<br>

##Solution Idea
The code is based on the idea of ​​making the Excel file entered into  the  SQL server database, row by row, with a period determined by the user of the code.<br>
Although the idea uses the Microsoft  SQL Server database, you will not need to write any query or code, You must only have the server installed on your device.
Then you create a new database in one step, which will be explained later in the steps.<br>


##Requirements
-Windows system has:<br>
- Power BI(Free plan).<br>
-Microsoft SQL Server Database (no coding experience needed).<br>


##Example Steps
The steps will be explained on the example of the dashboard attached in the first image, and you can apply the same steps to your own data and dashboard.

if you dont have, install Micrsoft SQL server and SQL Server Mangement Studio (SSMS). see only the first minute (1.12 minute) in this video [https://youtu.be/7zXtA0LwoHs?si=tWZjswcYrBLTY91Vt](https://youtu.be/7zXtA0LwoHs?si=tWZjswcYrBLTY91V)

###Step 3: Run the Python Code or The App
If you can deal with Python codes, go to the source code file "excel2rt.py".<br>
if you do not have a Python environment, you can use the Windows version of the code in the archive "[Excel To Real-Time.zip](https://drive.google.com/drive/folders/1SJVeZebtqcHyCtSfwGhgeID01J0fMzzi?usp=sharing)" extract it and then do the following steps:<br>
####Step 3.1: Copy and paste your Excel file  to the extracted folder
in this example I will use the "pizza_sales.xlsx" file, it's already in the zip archive.<br>
####Step 3.2: Run the App 'ExcelToRealTime.exe' that found in the extracted folder
####Step 3.3: The program will ask you to enter the name of the Excel file. In our example, I entered 'pizza_sales.xlsx'. after that click enter.<br>

####Step 3.4: The program will ask you to enter the time in second
enter a number bigger than 1, I entered 3, which means every new row will be added after 3 seconds.<br>

if you see the message "updating start" like this pic, that mean the Step 3 is done.<br>
app run.png

###Step 4: link your databse to the Power BI
